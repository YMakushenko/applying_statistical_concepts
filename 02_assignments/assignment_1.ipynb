### Question 1: Simple Linear Regression 

Let's set up our workspace and use the `Boston` dataset in the `ISLP` library. Print `Boston` to learn more about the dataset.

```python
# Import standard libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.pyplot import subplots
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Import specific objects
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

from statsmodels.stats.outliers_influence \
     import variance_inflation_factor as VIF
from statsmodels.stats.anova import anova_lm

from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
                         summarize,
                         poly)
```

```python
# Load the "Boston" dataset using the "load_data" function from the ISLP package
Boston = load_data('Boston')
```

Before we fit and review model outputs, we should visualize our data. Review the code and plot, shown below. Answer the following questions:

_(i)_ What are the `medv` and `dis` variables being plotted? (Hint: review this [link](https://islp.readthedocs.io/en/latest/datasets/Boston.html))

The 'medv' is for the median value of owner-occupied homes in $1000s.
The 'dis' is for the weighted mean of distances to five Boston employment centres.

_(ii)_ What concept ‘defines’ the plotted line?

The plotted line is a regression line. It's the best fit line through the data points. The concept that 'defines' the plotted line is a linear regression model that minimizes the sum of squared residuals (differences between observed and predicted values).

```python
# Extract the variables
medv = Boston['medv'].values.reshape(-1, 1)
dis = Boston['dis'].values.reshape(-1, 1)

# Plot data
plt.scatter(dis, medv, label='Data')
plt.xlabel('dis')
plt.ylabel('medv')

# Fit a linear regression model
lm = LinearRegression()
lm.fit(dis, medv)

# Plot the regression line
plt.plot(dis, lm.predict(dis), color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('dis')
plt.ylabel('medv')
plt.legend()

# Show the plot
plt.show()
```

Consider the variables plotted above. In the context of the `Boston` dataset:

_(iii)_ What is the (implied) null hypothesis? What is the (implied) alternative hypothesis?

The null hypothesis means that there is no linear relationship between 'dis' (distances to five Boston employment centres) and 'medv' (value of owner-occupied homes).
The alternative hypothesis means that there is a linear relationship between 'dis' and 'medv'.

_(iv)_ Now, let’s fit a simple regression model, using the general syntax `sm.OLS()` and `model.fit()`. As above, use `medv` as the response variable Y, and `dis` as the predictor variable X. (Hint: use the syntax `sm.add_constant` with the appropriate argument).

```python
import statsmodels.api as sm

# Use the response variable and add a constant to the predictor variable
X = sm.add_constant(dis)
y = medv

# Fit the simple regression model
model = sm.OLS(y, X).fit()
```

Review your model output to answer the following questions (Hint: use the `summary` and `conf_int` functions):    

```python
# Review the model summary
summary = model.summary()
print(summary)

# Get the confidence intervals
conf_int = model.conf_int()
print(conf_int)
```

_(v)_ What are the _coefficient estimates_ for $B_0$ (intercept) and $B_1$ (slope)?  

B₀ (intercept) = 18.3901.
B₁ (slope): 1.0916.

_(vi)_ What are the _standard errors_ for $B_0$ and $B_1$?  

Standart errors for B₀ = 0.817.
Standart errors for B₁ = 0.188.

_(vii)_ What are the _confidence intervals_ for $B_0$ and $B_1$?  

Confidence intervals for B₀ (intercept) = The 95% confidence interval is [16.784, 19.996].
Confidence intervals for B₁ (slope) = The 95% confidence interval is [0.722, 1.462].



Now, let's interpret the model output.  
_(viii)_ Is the model a good fit? (Hint: review $R^2$)  

R² = 0.062. It means that only 6.2% of the variability in the response variable 'dis' (median value of owner-occupied homes in $1000s) is explained by the predictor variable 'medv' (weighted mean of distances to five Boston employment centres). So the model may not be a good fit.

_(ix)_ Do we reject the (implied) null hypothesis? Why or why not? (Hint: review model $F$ statistic, $p$ value). 

F statistic = 33.58, p value = 1.21e-08.
We reject the (implied) null hypothesis because the p value for the slope is significantly less than 0.05. This means that there is a significant linear relationship between 'dis' and 'medv'.

### Question 2: Multiple Linear Regression 

We'll continue to use the `Boston` dataset for this question.

_(i)_ Fit a multiple linear regression, with two predictor variables: $X_1$ is `dis`, and $X_2$ is `rm`. As before, keep `medv` as the response variable Y. (Hint: use the syntax `sm.add_constant` with the appropriate arguments).

```python
# Extract additional variable 'rm'
rm = Boston['rm'].values.reshape(-1, 1)

# Combine 'dis' and 'rm' as predictors
X_mult = np.hstack((sm.add_constant(dis), rm))

# Fit the multiple regression model
model_mult = sm.OLS(y, X_mult).fit()
```

_(ii)_ In the context of the `Boston` dataset, state the null and alternative hypotheses.

The null hypothesis is that neither 'dis' nor 'rm' is linearly related to 'medv' (the coefficients for both 'dis' and 'rm' are zero).
The alternative hypothesis is that at least one of 'dis' or 'rm' is linearly related to 'medv' (at least one of the coefficients is not zero).

_(iii)_ Review the model output, using `summary()`. Does it appear that both `dis` and `rm` are predictive of `medv`? How did you determine this?

```python
# Review the model summary
summary_mult = model_mult.summary()
print(summary_mult)
```

From these results, both 'dis' and 'rm' are predictive of 'medv', as their p-values less than 0.05. The t-statistics are also well above the critical value for a typical significance level.

_(iv)_ We can use the inbuilt `sm.graphics.plot_regress_exog` function to generate helpful diagnostic plots (Hint: provide `plot_regress_exog` with the multiple regression model). Review the first generated plot, 'Residuals vs. Fitted'. Which observations are outliers? What impact might outliers have on our model?

```python
# Prepare the predictors
X = sm.add_constant(pd.DataFrame({'dis': dis, 'rm': rm}))

# Fit the multiple regression model
model_mult = sm.OLS(medv, X).fit()

# Generate diagnostic plots
fig = plt.figure(figsize=(12, 8))
sm.graphics.plot_regress_exog(model_mult, 'rm', fig=fig)
plt.show()
```

Outliears are those observations that lie far from the bulk of data points. They might have such an impact that it distorts the results and affects the estimation of the coefficients.

_(v)_ Fit a second model, this time including an interaction between the two predictor variables. Is there an interaction? (Hint: add a variable `x1 * x2` where `x1` and `x2` are the predictor variables). State an interpretation of the interaction, in the context of the `Boston` dataset, in one or two sentences.

```python
# Create interaction term
interaction = dis * rm

# Prepare the predictors including interaction
X_interaction = sm.add_constant(pd.DataFrame({'dis': dis, 'rm': rm, 'dis_rm': interaction}))

# Fit the model with interaction
model_interaction = sm.OLS(medv, X_interaction).fit()

# Review the model summary
print(model_interaction.summary())
```

The interaction has a positive value of 0.9855. I can state the following interpretation of it: the further the housing is located from employment centers, the stronger the cost of housing depends on the number of rooms; or, in other words, as the distance from employment center increases, the relationship between the number of rooms and the cost of housing becomes stronger.
